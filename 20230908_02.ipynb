```python
import pandas as pd

df = pd.read_csv('/Users/kimjuhwan/Desktop/data/wine/wine_data.txt', header=None)
df.columns=["wine","Alcohol","Malic acid","Ash","Alcalinity of ash","Magnesium","Total phenols","Flavanoids","Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315 of diluted wines","Proline"]
df
```

```python
df_data = pd.DataFrame(data=df, columns=["Malic acid","Alcalinity of ash","Flavanoids","Nonflavanoid phenols"])
df_target = pd.DataFrame(data=df, columns=["wine"])
```

```python
wine_data = pd.concat([df_data, df_target], axis=1)
```

```python
wine_data
```

```python
f, axes = plt.subplots(2,2,figsize=(8,6))
sns.distplot(wine_data['Malic acid'], ax=axes[0,0])
sns.distplot(wine_data['Alcalinity of ash'], ax=axes[0,1])
sns.distplot(wine_data['Flavanoids'], ax=axes[1,0])
sns.distplot(wine_data['Nonflavanoid phenols'], ax=axes[1,1])
plt.show()
```

```python
sns.pairplot(wine_data, vars=['Malic acid','Alcalinity of ash','Flavanoids','Nonflavanoid phenols'],
            hue='wine')
plt.show()
```

```python
colormap = plt.cm.PuBu
plt.figure(figsize=(8, 8))
plt.title("Analyze Correlation of Features", y=1.0, size =16)
sns.heatmap(wine_data.astype(float).corr(), linewidths =0.1, vmax=1.0,
           square=True, cmap=colormap, linecolor= "black", annot=True, annot_kws={"size":16})
```

```python
x = wine_data
x = x.drop(columns=['wine'], axis=1)
```

```python
x
```

```python
y= wine_data['wine']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# SVM
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

clf_svm = SVC(random_state=0)
clf_svm.fit(x_train, y_train)

pred_svm = clf_svm.predict(x_test)

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

```python
#LR

from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(x_train, y_train)

pred_lr = clf_lr.predict(x_test)

print("\n--- Logistic Regression Classifier ---")
print(accuracy_score(y_test, pred_lr))
print(confusion_matrix(y_test, pred_lr))
```

```python
#DT
from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(x_train, y_train)

pred_dt = clf_dt.predict(x_test)

print("\n--- Decision Tree Classifier ---")
print(accuracy_score(y_test, pred_dt))
print(confusion_matrix(y_test, pred_dt))
```

```python
#RT
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

print("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(x_train, y_train)
pred = rf_clf.predict(x_test)
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
```
